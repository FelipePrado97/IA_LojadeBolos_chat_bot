```

corpus = df[df.columns[:-1]].values
np = df[df.columns[1]].values 

vectorizer = TfidfVectorizer(sublinear_tf=True, max_df=0.5,strip_accents='unicode')

x = vectorizer.fit_transform(corpus.ravel())
y = np
ModelTree = DecisionTreeClassifier()
ModelKNN = KNeighborsClassifier(n_neighbors=1)
ModelRL = LogisticRegression(random_state=0, solver='lbfgs',multi_class='multinomial')

'''LeaveOneOut'''
from sklearn.model_selection import LeaveOneOut

loo = LeaveOneOut()
loo.get_n_splits(x)

outTree_predic_y = []
outTree_predic_y_true = []

outKN_predic_y = []
outKN_predic_y_true = []

outRL_predic_y = []
outRL_predic_y_true = []
for train_index, test_index in loo.split(x):

   #print("TRAIN:", train_index, "TEST:", test_index)
   x_train, x_test = x[train_index], x[test_index]
   y_train, y_test = y[train_index], y[test_index]
   #print(x_train, x_test, y_train, y_test)
   ModelTree = ModelTree.fit(x_train, y_train)
   ModelKNN = ModelKNN.fit(x_train, y_train)
   ModelRL = ModelRL.fit(x_train, y_train)
   
   outTree_predic_y.append(ModelTree.predict(x_test))
   outTree_predic_y_true.append(y_test)

   outKN_predic_y.append(ModelKNN.predict(x_test))
   outKN_predic_y_true.append(y_test)

   outRL_predic_y.append(ModelRL.predict(x_test))
   outRL_predic_y_true.append(y_test)

''' FIM LEAVEONEOUT'''



# Prediction accuracy
print("Accuracy for Decision Tree Classifier: " + str(accuracy_score(outTree_predic_y_true, outTree_predic_y)*100)+"%")
print("Accuracy for KNeighbors Classifier: " + str(accuracy_score(outKN_predic_y_true, outKN_predic_y)*100)+"%") 
print("Accuracy for Logistic Regression: " + str(accuracy_score(outRL_predic_y_true, outRL_predic_y)*100)+"%")
print("----------------------------")
# Prediction recall
print("Recall for Decision Tree Classifier: "+str(recall_score(outTree_predic_y_true, outTree_predic_y, average='macro')*100)+"%")
print("Recall for KNeighbors Classifier: "+str(recall_score(outKN_predic_y_true, outKN_predic_y, average='macro')*100)+"%")
print("Recall for Logistic Regression: "+str(recall_score(outRL_predic_y_true, outRL_predic_y, average='macro')*100)+"%")
print("-----------------------------")
# Prediction precision
print("Precision for Decision Tree Classifier: "+str(precision_score(outTree_predic_y_true, outTree_predic_y, average='macro')*100)+"%")
print("Precision for KNeighbors Classifier: "+str(precision_score(outKN_predic_y_true, outKN_predic_y, average='macro')*100)+"%")
print("Precision for Logistic Regression: "+str(precision_score(outRL_predic_y_true, outRL_predic_y, average='macro')*100)+"%")
print("-----------------------------")
```
